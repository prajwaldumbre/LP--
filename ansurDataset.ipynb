**Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
```

**Loading the Datasets (Male data and Female Data)**

```python
data_male = pd.read_csv("ANSUR_Male.csv",encoding='latin-1')
```

```python
data_male
```

```python
data_femal = pd.read_csv("ANSUR_Femal.csv")
```

```python
data_femal
```

**Analyzing Datasets**

```python
data_male.columns
```

```python
data_femal.columns
```

```python
data_femal.columns == data_male.columns
```

**Merging the Datasets**

```python
data = pd.concat([data_male,data_femal])
```

**Preprocessing the Data**

```python
data.columns[data.isnull().any()]
```

```python
data.Ethnicity = data.Ethnicity.fillna(data["Ethnicity"].mode().iloc[0])
```

```python
data
```

```python
data.columns[data.isnull().any()]
```

```python
data = data.drop("subjectid",axis = 1)
```

**Dividing data into numeric and Categorical Data**

```python
numeric_data = data.select_dtypes(int)
```

```python
object_data = data.select_dtypes(object)
```

```python
numeric_data
```

```python
object_data 
```

**Preprocessing the Categorical Data**

```python
object_data.Installation.unique()
```

```python
object_data.Component.unique()
```

```python
object_data.Branch.unique()
```

```python
object_data.PrimaryMOS.unique()
```

```python
len(object_data.SubjectsBirthLocation.unique())
```

```python
object_data.WritingPreference.unique()
```

```python
len(object_data.Ethnicity.unique())
```

```python
object_data["Gender"] = object_data["Gender"].apply(lambda x : 0 if x=="Female" else 1)
```

```python
ethnicity_fq = object_data.Ethnicity.value_counts()
```

```python
object_data.Ethnicity = object_data.Ethnicity.map(ethnicity_fq)
```

```python
oneHotWritingPreference = pd.get_dummies(object_data.WritingPreference).astype(int)
```

```python
object_data = pd.concat([object_data.drop("WritingPreference",axis = 1),oneHotWritingPreference] , axis=1)
```

```python
oneHotComponent = pd.get_dummies(object_data.Component).astype(int)
```

```python
object_data = pd.concat([object_data.drop("Component",axis = 1),oneHotComponent] , axis=1)
```

```python
oneHotBranch = pd.get_dummies(object_data.Branch).astype(int)
```

```python
object_data = pd.concat([object_data.drop("Branch",axis = 1),oneHotBranch] , axis=1)
```

```python
installation_fq =  object_data.Installation.value_counts()
object_data.Installation = object_data.Installation.map(installation_fq)
```

```python
subjectBirth_fq = object_data.SubjectsBirthLocation.value_counts()
object_data.SubjectsBirthLocation = object_data.SubjectsBirthLocation.map(subjectBirth_fq)
```

```python
primaryMOS_fq = object_data.PrimaryMOS.value_counts()
object_data.PrimaryMOS = object_data.PrimaryMOS.map(primaryMOS_fq)
```

```python
date_fq = object_data.Date.value_counts()
object_data.Date = object_data.Date.map(date_fq)
```

```python
object_data
```

**Merging Preprocessed Numeric and Categorical Data**

```python
data_main = pd.concat([numeric_data,object_data],axis=1)
```

```python
data_main
```

**Scalling the data fro PCA**

```python
sc = StandardScaler()
```

```python
scaled = sc.fit_transform(data_main)
```

```python
scaled_data = pd.DataFrame(scaled,columns=data_main.columns)
```

```python
scaled_data
```

```python
pca = PCA(n_components=10)
```

```python
pca_d = pca.fit_transform(scaled_data)
```

```python
print(pca.explained_variance_ratio_)
```

```python
pca_data  = pd.DataFrame(pca_d,columns=[f"component {i}" for i in range(10)])
```

```python
pca_data
```

**Analyzing Results of PCA**

```python

plt.figure(figsize=(15, 8))
plt.subplot(121)
sns.heatmap(data.select_dtypes(int,float).corr(),)
plt.title('Original Data\nCorrelation Matrix')

plt.subplot(122)
sns.heatmap(pca_data.corr())
plt.title('PCA Data\nCorrelation Matrix')


plt.tight_layout()
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(121)

plt.plot(range(1,11),pca.explained_variance_ratio_,'ro-')
plt.title("Individual Explained Varience ratios")
plt.xlabel("PCA components")
plt.ylabel("Explained Varience ratios")

plt.subplot(122)
plt.plot(range(1,11),np.cumsum(pca.explained_variance_ratio_),'bo-')
plt.title("Cumulative Explained Varience ratios")
plt.xlabel("PCA components")
plt.ylabel("Cumulative explained Varience ratios")
```

```python
print("\nExplained variance ratio by principal components:")

for i, ratio in enumerate(pca.explained_variance_ratio_, 1):
    print(f"PC{i}: {ratio:.4f} ({ratio*100:.2f}%)")
```
